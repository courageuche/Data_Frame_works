**<h2>To check the "DiscountOffer" cells with Rs</h2>**

```python
df['DiscountOffer'] = df['DiscountOffer'].astype(str)
data = df.loc[df['DiscountOffer'].str.contains('Rs')]
data 
```

**<h2>To check the 'DiscountOffer' cells with %</h2>

```python
df['DiscountOffer'] = df['DiscountOffer'].astype(str)
data1 = df.loc[df['DiscountOffer'].str.contains('%')]
data1
```

**<h2>Take the discountprice attached th 'Rs' in DiscountOffer Column to 'DiscountPrice' column</h2>**

```python
for i in range(len(data['DiscountOffer'])):
    arr = data['DiscountOffer'].values[i].split(" ")
    data['DiscountPrice (in Rs)'].values[i] = arr[1]
data
```

**<h2>Calculate the discountprice from the % discountOffer</h2>**

```python
for i in range(len(data1['DiscountOffer'])):
    arr = data1['DiscountOffer'].values[i].split('%')
    data1['DiscountPrice (in Rs)'].values[i] = data1['OriginalPrice (in Rs)'].values[i] - (int(arr[0]) * data1['OriginalPrice (in Rs)'].values[i]/100)
data1
```

**<h2>To show a dataframe consisting of both '%' and 'Rs' in DiscountOffer Column</h2>**

```python
data3 = df.loc[df['DiscountOffer'].str.contains('%') | df['DiscountOffer'].str.contains('Rs')]
data3
```

```python

data3['DiscountOffer'] = data3['DiscountOffer'].astype(str)
for i in range(len(data3['DiscountOffer'])):
    if 'Rs' in data3['DiscountOffer'].values[i]:
        arr_ = data3['DiscountOffer'].values[i].split(" ")
        data3['DiscountPrice (in Rs)'].values[i] = arr_[1]
    elif '%' in data3['DiscountOffer'].values[i]:
        arr_ = data3['DiscountOffer'].values[i].split('%')
        data3['DiscountPrice (in Rs)'].values[i] = data3['OriginalPrice (in Rs)'].values[i] - (int(arr_[0]) * data3['OriginalPrice (in Rs)'].values[i]/100)
data3
        
```

**<h2>Percentage of the original price that was discounted</h2>**

```python
_a = data3['DiscountPrice (in Rs)'].sum()
_b = data3['OriginalPrice (in Rs)'].sum()
print(_a)
print(_b)
```

```python
_c = _b - _a
_c
```

```python
_d = (_c/_b) * 100
_d
```

```python
np.round(_d, decimals = 2)
```
